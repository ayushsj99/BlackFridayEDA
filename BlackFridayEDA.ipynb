```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dftrain = pd.read_csv('Dataset/train.csv')
dftrain.head()
```

```python
dftest = pd.read_csv('Dataset/test.csv')
dftest.head()
```

```python
df = pd.concat([dftrain,dftest], axis=0, ignore_index=True)
df.head()
```

```python
df.info()
```

```python
df.drop(['User_ID'], axis=1, inplace=True)
```

```python
df.info()
```

```python
df['Gender'] = df['Gender'].map({'F':0, 'M':1})
df.head(10)
```

```python
df['Age'].unique()
```

```python
df['Age'] = df['Age'].map({'0-17':1, '18-25':2, '26-35':3, '36-45':4, '46-50':5, '51-55':6, '55+':7})
df.head(10)
```

```python
df['City_Category'].unique()
```

```python
df_city = pd.get_dummies(df['City_Category'], dtype=float)
df_city.head(10)
```

```python
df = pd.concat([df,df_city], axis=1)
df.head()
```

```python
df.drop('City_Category', axis=1, inplace=True)
df.head()
```

```python
df.isnull().sum()
```
